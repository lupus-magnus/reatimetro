```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
smooth10 = pd.Series(df.lin_signal).rolling(10).mean() #teoricamente fiz a suavização com média de 10 medidas.
smooth20 = pd.Series(df.lin_signal).rolling(20).mean()
smooth30 = pd.Series(df.lin_signal).rolling(30).mean()
smooth40 = pd.Series(df.lin_signal).rolling(40).mean()
smooth50 = pd.Series(df.lin_signal).rolling(50).mean()
```

```python
df['n10mean'] = smooth10
df['n20mean'] = smooth20
df['n30mean'] = smooth30
df['n40mean'] = smooth40
df['n50mean'] = smooth50
```

```python
#Daqui em diante, cuidarei do gráfico.
fig = plt.figure()
ax = fig.add_subplot()

x = df.time
y = df.n10mean
ax.plot(y,'black', label='mediamovel10', alpha = 1.0)
ax.plot(df.n20mean,'black', label='mediamovel20', alpha=0.75)
ax.plot(df.n30mean,'black', label='mediamovel30', alpha = 0.5)
ax.plot(df.n40mean,'black', label='mediamovel40', alpha = 0.25)
ax.plot(df.n50mean,'black', label='mediamovel50', alpha = 0.20)
ax.plot(df.lin_signal, 'r--', label='sinal linear cru')
ax.legend(loc='best') #adicionando a legenda
plt.xlim(725,1205) #Esse trecho e a linha abaixo servem somente para enfatizar a etapa de maior ruído da operação.
plt.ylim(630,880) #Dessa forma, posso olhar bem de perto o efeito da média móvel para diferentes n's nesse trecho.
```
